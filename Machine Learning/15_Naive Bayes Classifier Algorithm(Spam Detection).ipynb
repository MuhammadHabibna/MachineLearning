```python
import pandas as pd
```

```python
df = pd.read_csv('spam.csv')
df.head()
```

```python
df.groupby('Category').describe()
```

```python
df['spam'] = df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count, y_test)
#0.9883408071748879
```

# **sklearn pipeline**

> Menyederhanakan Basis Kode diatas



```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
#0.9883408071748879
```

## Gaussian Naive Bayes (Gaussian NB)

1. Data Type: Gaussian NB paling cocok untuk fitur yang bersifat kontinu, seperti data yang didistribusikan secara normal (Gaussian). Ini sering digunakan dalam kasus di mana fitur input berupa nilai-nilai numerik (misalnya, tinggi badan, berat badan, usia).

1. Assumption: Algoritma ini mengasumsikan bahwa setiap fitur mengikuti distribusi normal (Gaussian).

1. Example Use Case: Gaussian NB bisa digunakan untuk klasifikasi jenis kelamin berdasarkan ciri-ciri fisik (tinggi, berat), atau klasifikasi penyakit berdasarkan data biometrik seperti tekanan darah, kadar gula, dll.

1. **TIPS** : Gunakan Gaussian NB jika fitur-fitur yang Anda miliki berupa data numerik kontinu yang kira-kira mengikuti distribusi normal.



---
*Apa itu Distribusi Normal ?*

Distribusi normal adalah pola umum yang menggambarkan bagaimana data tersebar. Bentuknya seperti lonceng, dengan sebagian besar data berada di sekitar nilai tengah, dan semakin sedikit data yang ada di ujung-ujungnya.

Bayangkan kamu mengukur tinggi badan sekelompok besar orang. Sebagian besar orang akan memiliki tinggi yang mendekati rata-rata, sementara hanya sedikit orang yang sangat pendek atau sangat tinggi. Distribusi ini membentuk pola simetris, di mana rata-rata, median (nilai tengah), dan modus (nilai yang paling sering muncul) semuanya sama dan berada di tengah-tengah.

Distribusi normal sering digunakan dalam statistik karena banyak fenomena alamiah, seperti tinggi badan, nilai ujian, dan berat badan, cenderung mengikuti pola ini.

##Multinomial Naive Bayes (Multinomial NB)

1. Data Type: Multinomial NB lebih cocok untuk fitur yang bersifat diskrit atau berupa frekuensi atau jumlah hitungan (misalnya, jumlah kemunculan kata dalam dokumen). Algoritma ini banyak digunakan dalam klasifikasi teks.

1. Assumption: Multinomial NB mengasumsikan bahwa fitur adalah variabel diskrit, biasanya jumlah kemunculan (count) dari suatu kategori.

1. Example Use Case: Multinomial NB sering digunakan untuk klasifikasi teks seperti filtering spam, klasifikasi dokumen, atau analisis sentimen, di mana fitur yang digunakan adalah frekuensi kata atau kata-kata tertentu dalam teks.

1. **TIPS** : Gunakan Multinomial NB jika fitur-fitur Anda berupa data diskrit, terutama dalam analisis teks atau jika Anda memiliki data dalam bentuk frekuensi atau hitungan.



---

*Apa itu distribusi Diskrit*

Distribusi diskrit adalah jenis distribusi probabilitas yang digunakan untuk menggambarkan kejadian-kejadian yang memiliki jumlah hasil yang terbatas dan dapat dihitung.

**Karakteristik Distribusi Diskrit :**

1. Hasil yang Terbatas: Data hanya bisa berupa nilai-nilai tertentu (misalnya, angka bulat seperti 0, 1, 2, …).

1. Probabilitas Terpisah: Setiap nilai memiliki probabilitas tertentu yang terpisah. Misalnya, jika Anda melempar dadu, ada 6 kemungkinan hasil (1, 2, 3, 4, 5, atau 6), dan setiap hasil memiliki probabilitas 1/6.

1. Tidak Ada Nilai Di Antara: Antara dua nilai diskrit, tidak ada nilai lainnya. Misalnya, jika Anda menghitung jumlah orang dalam ruangan, nilainya bisa 10 atau 11, tapi tidak bisa 10.5.

**Contoh Distribusi Diskrit :**
1. Distribusi Binomial: Menggambarkan jumlah keberhasilan dalam serangkaian percobaan independen dengan dua hasil (misalnya, sukses atau gagal). Contoh: Jumlah kepala saat melempar koin 10 kali.

1. Distribusi Poisson: Menggambarkan jumlah kejadian dalam interval waktu atau ruang tertentu. Contoh: Jumlah panggilan telepon yang diterima oleh call center dalam satu jam.

1. Distribusi Multinomial: Perpanjangan dari distribusi binomial, digunakan untuk percobaan dengan lebih dari dua hasil. Contoh: Jumlah suara yang diterima oleh setiap kandidat dalam pemilu.
