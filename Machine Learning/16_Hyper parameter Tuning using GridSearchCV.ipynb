# **HyperParameter Tuning**

Proses pemilihan parameter Optimal



````python
from sklearn import svm, datasets
from sklearn.datasets import load_iris
from sklearn.model_selection import GridSearchCV
import pandas as pd
````

````python
iris = load_iris()
````

````python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df['flower'] = iris.target
df['flower'] = df['flower'].apply(lambda x: iris.target_names[x])
df.head()
````

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3)
````

````python
model = svm.SVC(kernel='rbf',C=30,gamma='auto')
model.fit(X_train,y_train)
model.score(X_test, y_test)
#0.9555555555555556
````

````python
from sklearn.model_selection import cross_val_score
````

````python
cross_val_score(svm.SVC(kernel='linear',C=10,gamma='auto'),iris.data, iris.target, cv=5)
````

````python
cross_val_score(svm.SVC(kernel='rbf',C=10,gamma='auto'),iris.data, iris.target, cv=5)
````

````python
cross_val_score(svm.SVC(kernel='rbf',C=20,gamma='auto'),iris.data, iris.target, cv=5)
````

# **BAGAIMANA CARANYA MENCARI PARAMETER YANG TERBAIK TANPA HARUS SATU-SATU ??**

> USE HYPERPARAMETER TUNING



````python
import numpy as np

kernels = ['linear', 'rbf']
C = [1,10,20]
avg_scores = {}
for kval in kernels:
    for cval in C:
        cv_scores = cross_val_score(svm.SVC(kernel=kval, C=cval, gamma='auto'), iris.data, iris.target, cv=5)
        avg_scores[kval + '_' + str(cval)] = np.average(cv_scores)

avg_scores
````



> **Masih tetap lama!**

Gunakan parameter Grid Search CV


````python
from sklearn.model_selection import GridSearchCV
````

````python
clf = GridSearchCV(svm.SVC(gamma='auto'), { #model dan parameter
    'C': [1,10,20],
    'kernel': ['rbf','linear']
}, cv=5, return_train_score=False
)

clf.fit(iris.data, iris.target)
clf.cv_results_ #dict of numpy(masked) ndarray
````

### *Rubah menjadi Data Frame*

````python
df = pd.DataFrame(clf.cv_results_)
df
````

````python
df[['param_C','param_kernel','mean_test_score']]
````

## **RandomizedSearchCV**

  tidak akan mencoba setiap permutasi dan kombinasi parameter, tetapi akan mencoba kombinasi acak dari nilai parameter

````python
from sklearn.model_selection import RandomizedSearchCV
````

````python
rs = RandomizedSearchCV(svm.SVC(gamma='auto'), {
    'C': [1,10,20],
    'kernel': ['linear', 'rbf']
}, cv=5, return_train_score=False, n_iter=2) #n_iter berapa kombinasi ?

rs.fit(iris.data, iris.target)
rs.cv_results_
pd.DataFrame(rs.cv_results_)[['param_C', 'param_kernel', 'mean_test_score']]
````

````python
# prompt: import svm, randomforest, logisticregression

from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

#manakah yang memiliki kinerja terbaik ????
````

````python
model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params' : {
            'C': [1,10,20],
            'kernel': ['rbf','linear']
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'n_estimators': [1,5,10]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params': {
            'C': [1,5,10]
        }
    }
}
````

````python
scores = []

for model_name, mp in model_params.items():
    clf =  GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(iris.data, iris.target)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
````

````python
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
````
