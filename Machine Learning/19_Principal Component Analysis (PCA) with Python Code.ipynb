### **PCA : Principal Component Anaysis**
 *Proses dimana mencari tahu features yang memiliki dampak maksimal pada suatu variabel target*

 Teknik statistik yang digunakan untuk mengurangi dimensi data sambil tetap mempertahankan sebanyak mungkin informasi yang ada. PCA bekerja dengan mengidentifikasi arah (komponen utama) di mana variasi data terbesar terjadi dan kemudian memproyeksikan data ke dalam ruang baru yang dibentuk oleh komponen-komponen ini.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
from sklearn.datasets import load_digits
digits = load_digits()
````

````python
df = pd.DataFrame(digits.data, columns=digits.feature_names)
df.head()
````

````python
X = df
y = digits.target
````

````python
X.shape
````

````python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.95) #pertahankan 95% features yang berguna
````

````python
X_pca = pca.fit_transform(X)
X_pca.shape #dari 64 features menjadi 29
````

````python
pca.explained_variance_ratio_ #kolom 1 menangkap 14% informasi berguna, seterusnya...)
````

````python
pca.n_components_ #kolom yang digunakan
````

````python
# prompt: import traintestsplit

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
````

````python
# prompt: import logistic regression

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
#.9611111111111111
#STOP: TOTAL NO. of ITERATIONS REACHED LIMIT. (harus pasang maks iter)
````

````python
model_iter = LogisticRegression(max_iter=1000)
model_iter.fit(X_train, y_train)
model_iter.score(X_test, y_test)
#0.9611111111111111
````

````python
pca = PCA(n_components=5) #merubah features menjadi 5 kolom
X_pca = pca.fit_transform(X)
X_pca.shape
#
````
