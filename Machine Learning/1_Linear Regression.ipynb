```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
canada = pd.read_csv('canada_per_capita_income.csv')
canada.head()
```

```python
X = canada[['year']]
y = canada['per capita income (US$)']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
regression.fit(X, y)
```

```python
prediction = regression.predict([[2020]]) # berapa target 2020?
prediction
```

```python
plt.plot(X, y, 'b.')
plt.plot(X, regression.predict(X), 'r-')
plt.show()
```
