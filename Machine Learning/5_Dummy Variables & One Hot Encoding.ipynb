# **ONE HOT ENCODING**

Membuat kolom baru untuk setiap kategori dan menetapkan nilai biner ( 1 / 0)

```python
import pandas as pd
```

```python
df = pd.read_csv('carprices.csv')
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merged = pd.concat([df, dummies], axis='columns')
merged
```

Menghapus kolum "car model" karena sudah tidak diperlukan

menghapus 1 jenis mobil agar tidak terjadi multi-ko-linear (jebakan biary)


```python
final = merged.drop(['Car Model', 'Audi A5'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop(['Sell Price($)'], axis='columns')
y = final['Sell Price($)']
```

```python
X
```

```python
y
```

```python
model.fit(X,y)
```

```python
model.predict([[45000,4,0,0]])
#luas, age, false,false (berarti selain mobil BMW dan Mercendez)
#kenapa ada 3 dimensi ? karena jenis mobil ada 3
```

```python
model.predict([[45000, 4, 0, 1]]) #mercendez
```

```python
model.predict([[86000,7, 1,0]]) #BMW
```

```python
model.score(X,y)
```

```python
le = LabelEncoder()
```

```python
X = dfle[['Car Model', 'Mileage', 'Age(yrs)']].values
X
```

```python
y = dfle['Sell Price($)'].values
y
```

# **sklearn ONE HOT ENCODER**

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(), [0])],
    remainder='passthrough'
)
```

```python
X = ct.fit_transform(X)
```

```python
X = X[:, 1:]
```

```python
model.fit(X,y)
```

```python
model.predict([[0,1,45000,4]])
```
