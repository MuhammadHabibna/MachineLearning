```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Cabin', 'Embarked'], axis=1, inplace=True)
df.head()
```

```python
# prompt: import label encoder

from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
# prompt: ticket label encoder

df['Ticket'] = le.fit_transform(df['Ticket'])
df.head()
#ternyta tiket berbeda semua
```

```python
df.drop(['Ticket'], axis=1, inplace=True)
df.head()
```

```python
dummies = pd.get_dummies(df.Sex)
dummies.head()
```

```python
df = pd.concat([df, dummies], axis='columns')
df.head()
```

```python
target = df.Survived
features = df.drop(['Survived'], axis='columns')
```

```python
target
```

```python
features
```

```python
features.Age = features.Age.fillna(features.Age.mean())
```

```python
features
```

```python
# prompt: import traintestsplit

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2)
```

```python
len(X_train)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.predict_proba(X_test[:10])
```
