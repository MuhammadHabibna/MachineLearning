```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
canada = pd.read_csv('canada_per_capita_income.csv')
canada.head()
```

```python
X = canada[['year']]
y = canada['per capita income (US$)']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[2020]])
```

# **Pickle**

1. keuntungan:
  1. Asli untuk Python
  1. Dapat membuat serial sebagian besar objek Python
  1. Penggunaan memori yang efisien dengan protokol=5

1. kekurangan :
  1. Pemuatan dapat mengeksekusi kode sembarangan
  1. Memerlukan lingkungan yang sama dengan lingkungan pelatihan



```python
import pickle
```

```python
with open('model_pickle', 'wb') as f: #write
  pickle.dump(model, f)  #menyimpan (men-serialisasi) objek Python ke dalam file
```

```python
with open('model_pickle', 'rb') as f: #read
  mp = pickle.load(f) #memanggil model
```

```python
mp.predict([[2020]])
```

# **JOBLIB**

1. Kelebihan
  1. Penggunaan memori yang efisien
  1. Mendukung pemetaan Memori
  1. Pintasan mudah untuk kompresi dan dekompresi

1. Kekurangan
  1. Format berbasis Pickle
  1. Pemuatan dapat mengeksekusi kode sembarangan
  1. Memerlukan lingkungan yang sama dengan lingkungan pelatihan
