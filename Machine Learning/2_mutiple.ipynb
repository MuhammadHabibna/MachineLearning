```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from word2number import w2n
```

```python
data = pd.read_csv('hiring.csv')
data
```

```python
data['experience'].fillna('zero', inplace=True)
data
```

```python
data['experience'] = data['experience'].apply(w2n.word_to_num)
data
```

```python
import math

median_score = math.floor(data['interview_score(out of 10)'].mean())
median_score
```

```python
data['test_score(out of 10)'].fillna(median_score, inplace=True)
data
```

```python
X = data[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = data['salary($)']
X
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X, y)
```

```python
prediction = reg.predict([[2, 10, 10]])
prediction
```
